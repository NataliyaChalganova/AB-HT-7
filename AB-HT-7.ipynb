#### Задание: На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem

#### Распишите, какие метрики и как вы будете анализировать

1. Если добавить фильтры для видео в функцию "кружочки" для всех пользователей, то это может увеличить среднее время использования на 5 %, потому что в чатах станут чаще использовать "кружочки".



1. Определение целевых метрик: среднее время использования функции "кружочки" в чатах.

2. Разделение на группы: Случайным образом, выберем две группы пользователей, контрольную группу (A) и тестовую группу (B). Обе группы должны быть представлены случайно выбранными пользователями, чтобы исключить возможные искажения.

Контрольная группа (A): Пользователи в этой группе останутся без изменений и будут продолжать использовать функцию "кружочки" без дополнительных фильтров.

Тестовая группа (B): Для пользователей в этой группе будет добавлен набор фильтров для видео в функцию "кружочки".

3. Измерение метрик: Соберем данные о среднем времени использования функции "кружочки" для обеих групп A и B в течение определенного периода времени (например, неделя).

4. Анализ данных: Для сравнения результатов между группами, проведем статистический анализ данных для проверки различий между группами.

5. Определение значимости: Сравним результаты и определим, есть ли статистически значимые различия в среднем времени использования функции "кружочки" между группами A и B. Если тестовая группа B показывает значительное и статистически значимое увеличение времени использования, то можно считать гипотезу подтвержденной.

6. Выводы: Исходя из результатов анализа, делаем выводы о том, как изменение функции "кружочки" влияет на среднее время использования в чатах. Если результаты эксперимента положительные, то эту измененную функцию можно внедрить для всех пользователей.

2. Если добавить двухфакторную аутентификацию для всех ползователей, это может увеличить количество активных пользователей на 5%, потому что станет больше доверия пользователей.

1. Определение целевых метрик: 
   - Количество активных пользователей
   - Конверсия зарегистрированных пользователей в активных
   - Время, проведенное пользователями в приложении
   - Оценка уровня доверия пользователей

2. Превентивные меры для защиты от "Peeking Problem":
   - Разделение пользователей на две группы: контрольную и тестовую
   - Тестирование двухфакторной аутентификации будет внедряться только для тестовой группы пользователей

3. Рандомизация выборки:
   - Рандомное распределение пользователей между контрольной и тестовой группами
   - Убедимся, что выборка подходит для обеих групп и является репрезентативной

4. Разработка и реализация функционала:
   - Создание функции двухфакторной аутентификации
   - Интеграция данной функции только для тестовой группы пользователей
   - Обеспечение стабильной работы и безопасности при использовании двухфакторной аутентификации

5. Запуск теста:
   - Запуск эксперимента на ограниченной выборке пользователей из тестовой группы
   - Собирание данных о результатах теста на протяжении определенного времени (например, 2-4 недели)

6. Анализ результатов:
   - Сравнение метрик активных пользователей в контрольной и тестовой группах
   - Проверка статистической значимости результатов с использованием соответствующих статистических методов
   - Оценка уровня доверия пользователей после внедрения двухфакторной аутентификации
